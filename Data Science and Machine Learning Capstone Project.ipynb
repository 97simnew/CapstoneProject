## Introduction

This document is an investigation using data analysis techniques and regression models completed in accordance with the IBM Data Science and Machine Learning Capstone Project.

This is an investigation into housing related complaints reported using the 311 number used by New York residents and how they were handled by New Yorks Department of Housing Preservation and Development (HPD).

In the city of New York, since 1978, the Department of Housing Preservation and Development has been responsible for preserving affordable housing, protecting tenants, and enforcing New Yorks Housing Maintenance Code. Today, tenants living in New York are able to report problems with properties they are renting using 311 non-emergency number. The HPD will then contact the property owner with the aim of ensuring a swift resolution of the complaint. An HPD representative may also visit the residence to assess its state, and can initiate legal action if it is unsatisfactory.

This investigation explores these datasets with the aim of helping the HPD to better understand the complaints which they are receiving, and respond to them effectively.

Questions examined will include:

1. What types of complaint do the HPD receive, and which of these should they prioritise?
2. Should the Department of HPD of New York City focus on any particular geographical areas where priority complaints are particularly numerous or severe?
3. Which building characteristics of New York residences make them most likely to be the subject of high priority HPD complaints?
4. Can a model be built to help the HPD predict future volumes priority complaint types?

### Changes in call rates over time
We will now examine how call frequency for different complaint types has changed over the last 10 years.

```python
complaint_types=['HEAT/HOT WATER','UNSANITARY CONDITION','APPLIANCE','GENERAL','ELECTRIC','PLUMBING','PAINT/PLASTER',
                 'WATER LEAK','SAFETY','DOOR/WINDOW','OUTSIDE BUILDING','ELEVATOR','FLOORING/STAIRS',
                 'HPD Literature Request','HEATING','GENERAL CONSTRUCTION','NONCONST']
year_dates=[pd.Timestamp(y,1,1) for y in range(2010,2021)]
Comps={}
for typ in complaint_types:
    Comps[typ]=complaints_df.loc[complaints_df['Complaint Type']==typ]
plt.figure(figsize=(12,6))
for key in Comps:
    val_10=Comps[key]['Created Date'].loc[Comps[key]['Created Date'].between(year_dates[0],year_dates[1])].size
    val_20=Comps[key]['Created Date'].loc[Comps[key]['Created Date'].between(year_dates[9],year_dates[10])].size
    print('%-25s2010: %-12s2020: %s' % (key,val_10,val_20))
    mean=Comps[key].shape[0]/10 #Mean number of each type of complaint per year. Used to normalise plots.
    y=[]
    x=list(range(2010,2020))
    for i in range(0,10):
        YearTotal=Comps[key]['Created Date'].loc[Comps[key]['Created Date'].between(year_dates[i],year_dates[i+1])].size
        y.append(((YearTotal/mean)-1)*100)
    plt.plot(x,y,label=key)
plt.xlabel('Year')
plt.legend(frameon=True,loc=(1.01,.27))
plt.ylabel('Frequency of complaint relative to that complaints average')
plt.title('Relative frequency of complaints 2010-2020')
plt.ylim(-100,170)
plt.xticks(range(2010,2020))
plt.show()
```

Looking at this graph, it is clear that a change in reporting practices occurred in 2014. HEATING, GENERAL CONSTRUCTION, and NONCONST were discontinued as categories. HEAT/HOT WATER, UNSANITARY CONDITION, WATER LEAK, SAFETY, DOOR/WINDOW, OUTSIDE BUILDING, ELEVATOR, and FLOORING/STAIRS, were introduced at about the same time. These changes clearly have a dramatic effect on the data, and so for now we shall focus analysis on the period since this change took place.

```python
new_list=['HEAT/HOT WATER','UNSANITARY CONDITION','WATER LEAK','SAFETY','DOOR/WINDOW','OUTSIDE BUILDING','ELEVATOR',
          'FLOORING/STAIRS']
old_list=['HEATING','GENERAL CONSTRUCTION','NONCONST']
new_df=pd.DataFrame(columns=['Complaint Type','Created Date'])
old_df=new_df.copy()
for key in new_list:
    new_df=new_df.append(Comps[key][['Complaint Type','Created Date']]
                         .sort_values('Created Date').head(1),ignore_index=True)
print('First instances of new Complaint Types:')
display(new_df)
for key in old_list:
    old_df=old_df.append(Comps[key][['Complaint Type','Created Date']]
                         .sort_values('Created Date',ascending=False).head(1),ignore_index=True)
print('Last instances of old Complaint Types:')
display(old_df)
```

We can see that the change in reporting system was introduced on 2014-03-11. Now we'll look at the annual numbers of complaints in different categories since this change was introduced.

```python
def rainbow_color(items,rots=3,light=.9):
    '''A function to assign colors to list of items. If rots=1 colors follow a rainbow pattern.
    rots changes the number of times the rainbow is cycled through to increase contrast.'''
    colors={}
    start=0
    done=0
    while done<len(items):
        for j in items[start::rots]:
            angleR=done*np.pi/len(items)
            angleG=angleR+2*np.pi/3
            angleB=angleG+2*np.pi/3
            colors[j]=((np.cos(angleR)**2)*light,(np.cos(angleG)**2)*light,(np.cos(angleB)**2)*light)
            done+=1
        start+=1
    sorted={}
    for i in items: sorted[i]=colors[i] #Restore original items order within dictionary
    return sorted
```

```python
complaint_types=['HEAT/HOT WATER','UNSANITARY CONDITION','APPLIANCE','GENERAL','ELECTRIC','PLUMBING','PAINT/PLASTER',
                 'WATER LEAK','SAFETY','DOOR/WINDOW','OUTSIDE BUILDING','ELEVATOR','FLOORING/STAIRS',
                 'HPD Literature Request']
Cdf_15_20=complaints_df.loc[(complaints_df['Complaint Type'].isin(complaint_types)) &
                            (complaints_df['Created Date'].dt.year>2014) &
                            (complaints_df['Created Date'].dt.year<2020)]
complaint_types.sort(key=lambda c: Cdf_15_20[['Complaint Type']].loc[(Cdf_15_20['Complaint Type']==c)].size,reverse=True)

color_dict=rainbow_color(complaint_types,rots=3,light=.85)
plt.figure(figsize=(8,7))
for year in range(2015,2020):
    year_df=Cdf_15_20.loc[(complaints_df['Complaint Type'].isin(complaint_types)) &
                            (complaints_df['Created Date'].dt.year==year)][['Complaint Type']]
    height=0
    for ct in complaint_types:
        
        count=year_df.loc[year_df['Complaint Type']==ct].count()
        plt.bar(year,count,bottom=height,color=color_dict[ct])
        height+=count
plt.xlabel('Year')
plt.legend(color_dict,frameon=True,loc=(1.01,.5))
plt.ylabel('Number of complaints by type')
plt.title('Complaints to 311 per year 2015-2019')
plt.show()
```

We can see that the general distribution of complaint types has remained fairly consistent over these 5 years. Though all complaints other than HEAT/HOT WATER seem to have declined somewhat in 2019. There is no way of telling from available data whether this is a real decline in complaints or just a delay in reporting.

### Seasonal variation

We will now look to see if there are any strong seasonal trends for different complaint types. For this analysis we shall again focus on complaints since 2015.

```python
complaint_types=['HEAT/HOT WATER','UNSANITARY CONDITION','APPLIANCE','GENERAL','ELECTRIC','PLUMBING','PAINT/PLASTER',
                 'WATER LEAK','SAFETY','DOOR/WINDOW','OUTSIDE BUILDING','ELEVATOR','FLOORING/STAIRS',
                 'HPD Literature Request']


Comps={}
for typ in complaint_types:
    Comps[typ]=complaints_df.loc[(complaints_df['Complaint Type']==typ) &
                                (complaints_df['Created Date'].between(pd.Timestamp(2015,1,1),pd.Timestamp(2020,1,1)))]

plt.figure(figsize=(12,6))
for key in Comps:
    mean=Comps[key].shape[0]/60 #Mean number of each type of complaint per month.
    y=[]
    x=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    for month in range(1,13):
        MonthTotal=Comps[key]['Created Date'].loc[Comps[key]['Created Date'].dt.month==month].size
        y.append(((MonthTotal/(5*mean))-1)*100)
    x.insert(0,0)
    x.append(13)
    y.insert(0,y[-1])
    y.append(y[1])
    plt.plot(x,y,label=key)
plt.xlabel('Month')
plt.legend(frameon=True,loc=(1.01,.4))
plt.ylim(-100,160)
plt.xlim(.5,12.5)
plt.ylabel('Frequency as percentage of 5 year average')
plt.title('Relative monthly frequency of complaints averaged for years 2015-2019')
plt.show()
```

Two clear seasonal trends can be seen here.
First, complaints about heating and hot water are far more common in the winter months (especially January) than in the summer months. Second, HPD Literature Requests are significantly higher in the spring than the autumn.

To understand how these changes may affect service demand, we'll re-plot these two together with total recorded complaints for the same period.

```python
complaint_types=['HEAT/HOT WATER','HPD Literature Request']

Comps={}
for typ in complaint_types:
    Comps[typ]=complaints_df.loc[(complaints_df['Complaint Type']==typ) &
                                (complaints_df['Created Date'].between(pd.Timestamp(2015,1,1),pd.Timestamp(2020,1,1)))]

Comps['Total Complaints']=complaints_df.loc[(complaints_df['Created Date'].dt.year>2014)&
                                            (complaints_df['Created Date'].dt.year<2020)]
    
plt.figure(figsize=(12,6))

for key in Comps:
    mean=Comps[key].shape[0]/60 #Mean number of each type of complaint per month.
    y=[]
    x=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    for month in range(1,13):
        MonthTotal=Comps[key]['Created Date'].loc[Comps[key]['Created Date'].dt.month==month].size
        y.append(((MonthTotal/(5*mean))-1)*100)
    x.insert(0,0)
    x.append(13)
    y.insert(0,y[-1])
    y.append(y[1])
    plt.plot(x,y,label=key)
plt.xlabel('Month')
plt.legend(frameon=True)
plt.ylim(-100,160)
plt.xlim(.5,12.5)
plt.ylabel('Frequency as percentage of 5 year average')
plt.title('Relative monthly frequency of seasonally varying and total complaints averaged for years 2015-2019')
plt.show()
```

The effect of seasonal variation in heating and hot water complaints on total service usage is very marked, due to the high proportion of total complaints which fall within this category. Literature requests however account for too small a proportion of total calls to have an impact on the overall distribution.

### Complaint resolution times

Finally, we will examine the the time taken to resolve complaints to see how quickly different types of complaint are handled, what affect seasonal variations in complaint volumes have on response times.

First let's look at the distribution of response times for different types of complaint since 2014.

```python
complaint_types=['HEAT/HOT WATER','UNSANITARY CONDITION','PAINT/PLASTER','PLUMBING','DOOR/WINDOW','WATER LEAK',
                 'ELECTRIC','GENERAL','FLOORING/STAIRS','APPLIANCE','SAFETY','HPD Literature Request',
                 'ELEVATOR','OUTSIDE BUILDING']
plt.figure(figsize=(12,8))
plot_df=Cdf_15_20.loc[(Cdf_15_20['Status']=='Closed')&
                      (Cdf_15_20['Closed Date'].dt.year>2014)].copy()
plot_df.loc[:,'Duration']=plot_df['Closed Date']-plot_df['Created Date']
plot_df.loc[:,'Duration']=plot_df['Duration'].apply(lambda i:int(i.days))
X=[]
X.append(plot_df['Duration'])
for ct in complaint_types:
    X.append(plot_df['Duration'].loc[plot_df['Complaint Type']==ct])
complaint_types.insert(0,'Total')
plt.boxplot(X[::-1],vert=False,labels=complaint_types[::-1],showmeans=True,showfliers=False)
plt.xlim(left=0)
plt.xlabel('Days')
plt.title('Time taken to close complaints by type 2015-2019')
plt.show()
```

Many complaints are clearly handled very quickly, with more than 50% of all complaints are closed in 5 days or less. However as 25% of all complaints take more than 2 weeks to close, it is clear that a significant proportion of cases are much slower to resolve.

Heating/Hot water complaints are clearly prioritised and have far shorter response times than most other types of complaints with 75% of complaints resolved in 4 days or less. HPD literature requests, also tend to be closed quickly, though this is probably because they are much simpler to respond to than other types of complaint.

Next we will examine the effect that seasonal variation in Heating/Hot water complaints has on complaint response times.

```python
plt.figure(figsize=(12,6))
plot_df=Cdf_15_20.loc[((Cdf_15_20['Status']=='Closed')&
                      (Cdf_15_20['Complaint Type']=='HEAT/HOT WATER'))&
                      (Cdf_15_20['Closed Date']>pd.Timestamp(2015,1,1))].copy()
plot_df.loc[:,'Duration']=plot_df['Closed Date']-plot_df['Created Date']
plot_df.loc[:,'Duration']=plot_df['Duration'].apply(lambda i:int(i.days))
labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
X=[]
for i in range(1,13):
    X.append(plot_df['Duration'].loc[plot_df['Created Date'].dt.month==i])
plt.boxplot(X,vert=True,labels=labels,showmeans=True,showfliers=False)
plt.ylim(bottom=-.05)
plt.ylabel('Days')
plt.title('Seasonal times taken to close a HEAT/HOT WATER complaints 2015-2019')
plt.show()
```

As the figure shows, response times for Heating/Hot water complaints are low throughout the year, and don't show any particular rise during the winter, when complaint volume increases. In fact, response times seem to be longest during the summer (possibly because a higher proportion of complaints received arise from complicated or unusual problems).

It seems that the HPD have strategies for handling the volume of these complaints in the winter, which prevent response times rising for this type of complaint. However, if we examine seasonal response times for other types of complaints a more complex picture is revealed.

### Is the distribution of complaints between zip codes consistent with time?

This section will investigate the type and degree of correlation between complaint volume within each zip code, and year between 2010 and 2020. To investigate this question, we must first drop all records which have no Incident Zip recorded. This amounts to 19024 heating/hot water complaint records (0.89% of total).

We will begin by looking at how variable the number of cases within zip codes is by calculating the average standard deviation and and coefficient of variation.

```python
zips=Q2_df['Incident Zip'].dropna().unique()
date_df=Q2_df[['Incident Zip','Created Date']].copy().dropna(subset=['Incident Zip'])
date_df=date_df.loc[date_df['Created Date'].dt.year.between(2010,2019,inclusive=True)]
date_df['Created Year']=date_df['Created Date'].dt.year
date_df=date_df.groupby(['Incident Zip','Created Year']).count()
date_df.rename(columns={'Created Date':'Complaint volume'},inplace=True)

for z in zips:
    for y in range(2010,2020):
        if (z,y) not in date_df.index:
            date_df.loc[(z,y),'Complaint volume']=0

ZipFit_df=pd.DataFrame(columns=['ZipCode','Variation','Std error','Std dev','Mean']).set_index('ZipCode')
years=[y for y in range(2010,2020)]
for z in zips:
    vols=date_df['Complaint volume'].loc[z].loc[years].tolist()
    s,i,r,p,err=stats.linregress(x=years,y=vols)
    ZipFit_df.loc[z,'Variation']=stats.variation(vols)
    ZipFit_df.loc[z,'Std error']=err
    ZipFit_df.loc[z,'Std dev']=np.std(vols)
    ZipFit_df.loc[z,'Mean']=np.mean(vols)
    
print('Average standard deviation:',np.mean(ZipFit_df['Std dev']))
print('Average coefficient of variation:',np.mean(ZipFit_df['Variation']))
```

Looking at these numbers, we can see that there is considerable variation in the volume of complaints from particular zip codes from year to year. The average standard deviation is 172, and the average coefficient of variation is 0.43. This means that the average standard deviation in complaint volume is 43% of the mean. Clearly complaint volume within zip codes is far from being constant with time. However, if we try applying a linear model to the change in call volumes from year to year, results are more promising.

```python
err=np.mean(ZipFit_df['Std error']/ZipFit_df['Mean'])
print('Average relative standard error for linear fits:',err)
```

The average relative standard error for simple linear models is 0.046. This is only a preliminary result, but it does suggest that annual complaint volume in many zip codes may be amenable to a linear model. This is our first indication that the heating and hot water complaints my be dependent on the year, as well as the time of year. This result is very preliminary, without data preprocessing or separate test data. A far more thorough model will be developed in section 6.

## A predictive model for heating and hot water complaints

### Predicting complaint volume from Month, Year, and Zip-code

This model will attempt to predict complaint volume from a zip code during a given month. Dummies will be created for zip codes, so that these are treated as independent variables rather than a continuum. A polynomial transform will be used for month, as this clearly doesn't have a linear relationship with complaint volume. A cubic transform will be used as this a sine-curve like pattern, as would be expected for a cyclical monthly variation.

```python
import sklearn.preprocessing as prep
model_df=Q2_df[['Incident Zip','Created Date']].dropna()
print('There are',Q2_df.shape[0],'complaint records.')
print (Q2_df.shape[0]-model_df.shape[0],'of these records are dropped as they lack either zip code, or date.')

YearScale=prep.StandardScaler().fit(np.array(model_df['Created Date'].dt.year).reshape(-1,1))
MonthScale=prep.StandardScaler().fit(np.array(model_df['Created Date'].dt.month).reshape(-1,1))
Poly=prep.PolynomialFeatures(degree=3,include_bias=False)
def prepare(df):
    df=df.join(pd.get_dummies(df['Incident Zip'],prefix='z'))
    df=df.drop(labels=['Incident Zip'],axis=1)
    
    df['Year']=YearScale.transform(np.array(df['Year']).reshape(-1,1))
    df['Month']=MonthScale.transform(np.array(df['Month']).reshape(-1,1))
    
    temp_df=df.loc[:,~df.columns.isin(['Month'])]
    df=temp_df.join(pd.DataFrame(Poly.fit_transform(df[['Month']]),columns=['Month'+str(n) for n in range(1,4)]))
    return df

test_size=.2
test_df=model_df.sample(frac=test_size,replace=False,random_state=0,axis=0)
test_rows=[i for i in model_df.index if i not in test_df.index]
train_df=model_df.loc[test_rows,:]

zips=model_df['Incident Zip'].unique()
z_names=['z_'+str(z) for z in zips]
def zeros(df,years=range(2010,2020)):
    for z in zips:
        for y in years:
            for m in range(1,13):
                if df.loc[(df['Incident Zip']==z)&((df['Year']==y)&(df['Month']==m))].size==0:
                    df=df.append({'Incident Zip':z,'Year':y,'Month':m,'Count':0},ignore_index=True)
    return df

test_df['Month']=test_df['Created Date'].dt.month
test_df['Year']=test_df['Created Date'].dt.year
cols=[c for c in test_df.columns if c!='Created Date']
test_df=test_df.groupby(cols).count().rename(columns={'Created Date':'Count'})
test_df['Count']=test_df['Count']/test_size  #scale up to propertional full count for 1 year.
test_df.reset_index(inplace=True)
test_df=zeros(test_df)
test_df=prepare(test_df)

train_df['Month']=train_df['Created Date'].dt.month
train_df['Year']=train_df['Created Date'].dt.year
train_df=train_df.groupby(cols).count().rename(columns={'Created Date':'Count'})
train_df['Count']=train_df['Count']/(1-test_size)  #scale up to propertional full count for 1 year.
train_df.reset_index(inplace=True)
train_df=zeros(train_df)
train_df=prepare(train_df)

#Need to rename columns, as '.0' confuses patsy in comand strings
def rename(c):
    if c in z_names: return str(c)[:-2]
    else: return c
train_df.rename(columns=rename,inplace=True)
test_df.rename(columns=rename,inplace=True)

print('Test:',test_df.shape,'Train:',train_df.shape)
train_df.head()
```

```python
from sklearn.linear_model import LinearRegression as LinReg
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from patsy import dmatrices

model1=RidgeCV(alphas=[0.0001,0.001,0.01,0.1,1,10,100],store_cv_values=True)
#FormStr tells the dmatrices function to return interaction terms between zip-code dummies, and date.
FormStr=('Count ~ ('+' + '.join(z_names)+') : (Year + Month1 + Month2 + Month3)').replace('.0','')
y_train,X_train=dmatrices(FormStr,train_df)
y_test,X_test=dmatrices(FormStr,test_df)
model1.fit(X_train,y_train)

print('Alpha value selected by cross-validation:',model1.alpha_)
print('R^2 score:',model1.score(X_test,y_test))
```

After tuning the parameter alpha (which affects the weight given to outliers), our linear model, trained with 80% of all complaints achieves an R-squared score of 0.85 for our test data. In other words it is able to account for 85% of the variability within that data. This is good accuracy score, and suggests that the variables used are good predictors of complaint volume. The alpha value of 1 selected by cross validation here will be used for further implementations of this model.

So far however, we have only tested the ability of this model to predict complaints in the period 2010-2020, based on data about complaints during the period 2010-2020. To be of use to the HPD, this model must be able to extrapolate from past complaints to predict future ones. We will therefore train and test our model again, using all data prior to the start of 2019 as training data, and all data from 2019 as test data. This will give an indication of our models ability to predict complaint volumes based on past trends.

```python
test_df=model_df.loc[model_df['Created Date'].dt.year==2019].copy()
train_df=model_df.loc[model_df['Created Date'].dt.year<2019].copy()

test_df['Month']=test_df['Created Date'].dt.month
test_df['Year']=test_df['Created Date'].dt.year
cols=[c for c in test_df.columns if c!='Created Date']
test_df=test_df.groupby(cols).count().rename(columns={'Created Date':'Count'})
test_df['Count']=test_df['Count']
test_df.reset_index(inplace=True)
test_df=zeros(test_df,years=[2019])
test_df=prepare(test_df)

train_df['Month']=train_df['Created Date'].dt.month
train_df['Year']=train_df['Created Date'].dt.year
train_df=train_df.groupby(cols).count().rename(columns={'Created Date':'Count'})
train_df['Count']=train_df['Count']
train_df.reset_index(inplace=True)
train_df=zeros(train_df,years=[y for y in range(2010,2019)])
train_df=prepare(train_df)

train_df.rename(columns=rename,inplace=True)
test_df.rename(columns=rename,inplace=True)

print('Test:',test_df.shape,'Train:',train_df.shape)

y_train,X_train=dmatrices(FormStr,train_df)
y_test,X_test=dmatrices(FormStr,test_df)
model1=Ridge(alpha=1,tol=1e-21).fit(X_train,y_train)
print('R^2 score for 2019 complaint predictions:',model1.score(X_test,y_test))
```

Making predictions in this way our model achieves an R-squared score of 0.88 It performs even better than in the previous test (possibly because more of the data was available as training data). This result is reassuring, and suggests that this kind of linear model could be a useful, and fairly reliable, predictive tool for the HPD.

### How model predictions compare to reality

This section will plot a number of graphs to compare how our predictive model compares to past results, and look at the predictions it makes for the future. The aim is to better understand the behavior of this model, and determine its strengths and weaknesses as a predictive tool. To do this, we will first retrain the model using all data from 2010-2020.

```python
train_df=model_df.loc[model_df['Created Date'].dt.year<2020].copy()

train_df['Month']=train_df['Created Date'].dt.month
train_df['Year']=train_df['Created Date'].dt.year
cols=[c for c in train_df.columns if c!='Created Date']
train_df=train_df.groupby(cols).count().rename(columns={'Created Date':'Count'})
train_df['Count']=train_df['Count']
train_df.reset_index(inplace=True)
train_df=zeros(train_df,years=[y for y in range(2010,2019)])
train_df=prepare(train_df)

train_df.rename(columns=rename,inplace=True)

y_train,X_train=dmatrices(FormStr,train_df)
model1=Ridge(alpha=1,tol=1e-21).fit(X_train,y_train)
print('Model trained from all complaints records 2010-2020')
```

First, we'll compare predictions of this model to complaint records for 4 randomly selected zip codes.

```python
zips=np.random.choice(model_df['Incident Zip'].unique(),size=4,replace=False)
months=[m for m in range(1,145)]
fig,axs=plt.subplots(4,1,figsize=(12,18))
fig.tight_layout(pad=5)
def YM(m):
    Y=int((m-1)/12)
    M=m-(12*Y)
    return (Y+2010),M
for i,z in enumerate(zips):
    ys=[]
    raw_X_df=pd.DataFrame(columns=train_df.columns)
    for m in months:
        Y,M=YM(m)
        y=sum(((model_df['Created Date'].dt.year==Y)&
               (model_df['Created Date'].dt.month==M))&
              (model_df['Incident Zip']==z))
        ys.append(y)
        raw_X_df.loc[m-1,'Year']=YearScale.transform([[Y]])[0,0]
        raw_X_df.loc[m-1,('z_'+str(z)[:-2])]=1
        raw_X_df.loc[m-1,'Month1':'Month3']=Poly.fit_transform(MonthScale.transform([[M]]))[0]
    raw_X_df.fillna(0,inplace=True)
    y,Xs=dmatrices(FormStr,raw_X_df)
    y_hats=model1.predict(Xs)
    axs[i].plot(months[:120],ys[:120],color='tab:blue',label='Recorded complaint volume')
    axs[i].plot(months,y_hats,color='tab:red',label='Predicted complaint volume')
    axs[i].legend()
    axs[i].set(title='Model prediction and records for heating complaints in '+str(int(z)),
               ylabel='Monthly complaint volume',
               xlabel='Year',
               xlim=(0,144))
    axs[i].set_xticks([(1+12*y) for y in range(0,12)])
    axs[i].set_xticklabels([y for y in range(2010,2022)])
    plt.setp(axs[i].xaxis.get_majorticklabels(),ha='left')

plt.show()
```

As we can see, this model achieves a good level of fit for zip-codes with high complaint volumes. It is less reliable however for zip-codes with low complaint volumes. This is because the absolute error in these zip-codes remains small, and is therefore within tolerance, despite being a large error relative to the complaint volumes for these zip-codes. To reduce this issue, the tolerance was reduced to 10^-21. However this has not eliminated the problem entirely. Fortunately, it is predicting high local complaint volumes which is particularly important to the HPD for future planning, and this model is far more successful at predicting these.

Next we should check whether this model is successful in predicting complaint volumes for New York as a whole, as well as individual zip-codes.

```python
zips=model_df['Incident Zip'].unique()
months=[m for m in range(1,145)]
ys=[]
y_hats=[]
plt.figure(figsize=(14,5))
for m in months:
    Y,M=YM(m)
    y=sum((model_df['Created Date'].dt.year==Y)&
          (model_df['Created Date'].dt.month==M))
    ys.append(y)
    
    raw_X_df=pd.DataFrame(columns=train_df.columns)
    for i,z in enumerate(zips):
        raw_X_df.loc[i,'Year']=YearScale.transform([[Y]])[0,0]
        raw_X_df.loc[i,('z_'+str(z)[:-2])]=1
        raw_X_df.loc[i,'Month1':'Month3']=Poly.fit_transform(MonthScale.transform([[M]]))[0]
    raw_X_df.fillna(0,inplace=True)
    y,Xs=dmatrices(FormStr,raw_X_df)
    y_hats.append(sum(model1.predict(Xs)))
plt.plot(months[:120],ys[:120],color='tab:blue',label='Recorded complaint volume')
plt.plot(months,y_hats,color='tab:red',label='Predicted complaint volume')
plt.title('Model prediction and records for total heating complaints in New York')
plt.ylabel('Monthly complaint volume')
plt.xlabel('Year')
plt.xlim(0,144)
plt.legend()
plt.xticks([(1+12*y) for y in range(0,12)],[y for y in range(2010,2022)],ha='left')

plt.show()
```

Here we can see that although our model is not fitted as well as might be wished for zip-codes with low complaint volumes, it is a very good fit for New York as a whole, as well as zip-codes with higher complaint volumes. This suggests that it would prove an effective predictive tool for any area with a reasonably high total heating and hot water complaint volume.

### Conclusions

In this section, a regression model has been developed with the aim of predicting the future volume of heating and hot water complaints from different New York zip codes. Important features of the developed model are:

- The variables used to predict future volume are Month, Year, and Zip-code.
- Year is treated as a linear variable. Month is used in a cubic polynomial transform to allow for cyclical behavior. Zip-code is treated as a categorical variable.
- Interaction terms are found for each zip-code, with month and year variables to allow for different patterns within each zip-code.
- A ridge regression model is used with a value of alpha=1 determined by cross-validation testing.
- To improve fitting for zip-codes with low complaint volumes, the fit tolerance was reduced to 10^-21.
- This model, trained on data for Jan 2010 - Dec 2018, achieves an R-squared score of 0.883 for predicting complaint volumes in 2019.
- Strengths of this model: Achieves a good accuracy when predicting future complaint volumes. Predictions can be made based on simple and readily available information.
- Weaknesses of this model: Fit is still much better for zip-codes with high call volume, than those with low volumes. Relies on extrapolating from recent trends, and may be unreliable in the event of new demographic changes.

An alternative approach to this problem would be to model the probable complaint volume within a zip-code based on the type of properties found in that zip-code. The advantages of such a model would be that it could predict changes in complaint volumes caused by changing demographics, and attempt predictions for zip-codes for which prior records were unavailable. On the other hand, as we saw in section 5, correlations between the property features we can establish and complaint volumes are not particularly strong. Difficulties of connecting recorded complaints to property records would also pose difficulties for both developing and using such a model.

To improve this situation, the HPD might consider hiring additional seasonal workers in the winter when heating and hot water complaints are frequent. A model of this seasonal variation might also help them to match their resources and manpower to this seasonal variation.

Complaint volumes have been largely consistent year on year for boroughs, and although complaint volumes vary within zip codes, they follow a roughly linear pattern with year. This suggests that a linear model might be a reliable method to extrapolate complaint volumes from different zip codes in future years.

It was established that the majority of complaints relate to residential buildings holding multiple families which were built before 1940. The HPD Might therefore wish to tailor training for their staff, to focus on issues often found in large buildings from this era.

Finally, section 6 developed a model to predict future heating and hot water complaint volume for different zip-codes in future months. Using 2019 records as test data, this model achieved an R-squared score of 0.88, indicating that the model is effective at extrapolating to future complaint volumes. With this model, the HPD would be able to anticipate which areas will have the highest heating and hot water complaint volumes in the coming months, and allocate resources to deal with these complaints as effectively as possible.
